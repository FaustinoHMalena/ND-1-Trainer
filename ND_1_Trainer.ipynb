```python
import json

SYS_ND = (
  "You are ND-1, a calm, evidence-based companion for neurodivergent minds. "
  "You teach, guide, and help users explore whether their experiences may align "
  "with autism, ADHD, dyslexia, OCD or related profiles — without ever diagnosing. "
  "Adapt depth and tone to the user. Be honest, concrete, and never condescending."
)

def _msgs(user, assistant, system=SYS_ND):
    u, a = str(user or "").strip(), str(assistant or "").strip()
    if not u or not a: return None
    return {"messages":[{"role":"system","content":system},
                        {"role":"user","content":u},
                        {"role":"assistant","content":a}]}

# --- screening adapter helpers ---
_SCREEN_LABEL_KEYS = ("Class","Class/ASD","class","label","Label","ASD","asd","diagnosis","Diagnosis","outcome")
def _screen_label(r):
    for k in _SCREEN_LABEL_KEYS:
        if k in r and r[k] not in (None,""):
            v = str(r[k]).strip().lower()
            if v in ("yes","1","true","positive","asd","autism"): return "indicators consistent with autism are present"
            if v in ("no","0","false","negative","control"):       return "indicators are below the screening threshold"
            return f"screening result: {v}"
    return ""

# --- transcript adapter helper ---
_TRANSCRIPT_KEYS = ("text","transcript","content","body","article","chunk","passage")
def _transcript_text(r):
    for k in _TRANSCRIPT_KEYS:
        if r.get(k): return str(r[k])[:1500]
    return json.dumps(r)[:1500]

ADAPTERS = {
 "instruct_io":   lambda r: _msgs(r.get("instruction") or r.get("prompt") or r.get("input",""),
                                  r.get("output") or r.get("response") or r.get("completion") or r.get("answer","")),
 "qa_pair":       lambda r: _msgs(r.get("question") or r.get("Context") or r.get("input") or r.get("query",""),
                                  r.get("answer")   or r.get("Response")or r.get("output","")),
 "ctx_response":  lambda r: _msgs(r.get("Context",""), r.get("Response","")),
 "qa_cot":        lambda r: _msgs(r.get("question") or r.get("problem") or r.get("prompt",""),
                                  r.get("answer")   or r.get("solution") or r.get("response","")),
 "mcq":           lambda r: _msgs(
        f"{r.get('question') or r.get('question_stem','')}\nOptions: "
        f"{r.get('choices') or r.get('options') or r.get('endings','')}",
        str(r.get('answerKey') or r.get('answer') or r.get('label',''))),
 "hellaswag":     lambda r: _msgs(
        f"{r.get('ctx','')}\nEndings: {r.get('endings','')}",
        str(r.get('label',''))),
 "classify":      lambda r: _msgs(
        f"Classify: {r.get('text') or r.get('headline') or r.get('tweet') or r.get('post','')}",
        str(r.get('label') or r.get('class') or r.get('disorder',''))),
 "emotions":      lambda r: _msgs(
        f"What emotion is expressed here?\n{r.get('text') or r.get('sentence','')}",
        str(r.get('label_text') or r.get('emotion') or r.get('label',''))),
 "summarize":     lambda r: _msgs(
        f"Summarize concisely:\n{r.get('document') or r.get('article') or r.get('dialogue') or r.get('text','')}",
        r.get('summary') or r.get('highlights') or r.get('abstract','')),
 "raw_text":      lambda r: _msgs("Reframe this for a neurodivergent reader in plain, supportive language:",
                                  (r.get('text') or r.get('description') or r.get('content') or r.get('body') or json.dumps(r))[:1500]),
 "dialog":        lambda r: _msgs(r.get('input') or r.get('context') or r.get('previous_utterance',''),
                                  r.get('response') or r.get('target') or r.get('free_messages','')),
 "preference_sft":lambda r: _msgs(r.get('prompt') or r.get('instruction') or r.get('input',''),
                                  (r.get('chosen') or [{}])[-1].get('content','')
                                  if isinstance(r.get('chosen'),list)
                                  else (r.get('chosen') or r.get('response',''))),
 "nectar":        lambda r: _msgs(r.get('prompt',''),
                                  (r.get('answers') or [{}])[0].get('answer','') if r.get('answers') else ''),
 "aya":           lambda r: _msgs(r.get('inputs',''), r.get('targets','')),

 # --- NEW ---
 "transcript":    lambda r: _msgs(
        "Summarize this neuroscience/clinical talk for a neurodivergent learner in plain language:",
        _transcript_text(r)),
 "screening":     lambda r: _msgs(
        "Based on these screening responses, what is the likely indication?\n"
        + json.dumps({k:v for k,v in r.items() if k not in _SCREEN_LABEL_KEYS})[:1200],
        _screen_label(r)),
}
```

```python
from datasets import load_dataset, Dataset, concatenate_datasets

PER_DATASET_CAP = 400  # match what your probe used

def take_sampled(hf_id, adapter, cfg, cap=PER_DATASET_CAP):
    try:
        ds = load_dataset(hf_id, cfg, split="train", streaming=True)
    except Exception as e:
        print(f"[skip] {hf_id} ({cfg}): {e}")
        return None
    rows, fn = [], ADAPTERS[adapter]
    for i, ex in enumerate(ds):
        if i >= cap * 6: break
        try:
            m = fn(ex)
            u = m["messages"][1]["content"]
            a = m["messages"][2]["content"]
            if u and a:
                rows.append(m)
        except Exception:
            pass
        if len(rows) >= cap: break
    print(f"[ok]   {hf_id}: {len(rows)} rows")
    return Dataset.from_list(rows) if rows else None

parts = []
for layer, hf_id, adapter, cfg in REGISTRY:
    d = take_sampled(hf_id, adapter, cfg)
    if d is not None:
        parts.append(d)

train_ds = concatenate_datasets(parts).shuffle(seed=42)
print("\n=== TOTAL TRAINING ROWS:", len(train_ds), "===")
print("Columns:", train_ds.column_names)
```

```python
def fmt(batch):
    texts = [tokenizer.apply_chat_template(m, tokenize=False, add_generation_prompt=False)
             for m in batch["messages"]]
    return {"text": texts}

train_ds = train_ds.map(fmt, batched=True, remove_columns=train_ds.column_names)
print("Formatted rows:", len(train_ds))
print("Sample:\n", train_ds[0]["text"][:500])

```

```python
from trl import SFTTrainer, SFTConfig
from unsloth.chat_templates import train_on_responses_only


trainer = SFTTrainer(
    model         = model,
    tokenizer     = tokenizer,
    train_dataset = train_ds,
    dataset_text_field = "text",
    max_seq_length = MAX_SEQ_LEN,
    packing       = True,
    args = SFTConfig(
        output_dir              = OUTPUT_DIR,
        per_device_train_batch_size = 8,
        gradient_accumulation_steps = 4,
        warmup_ratio    = 0.03,
        num_train_epochs= 1,
        learning_rate   = 5e-5,
        logging_steps   = 25,
        save_steps      = 500,
        bf16            = torch.cuda.is_bf16_supported(),
        fp16            = not torch.cuda.is_bf16_supported(),
        optim           = "adamw_8bit",
        lr_scheduler_type = "cosine",
        seed            = 42,
        report_to       = "none",
),
)


# Mask loss on user turns; only train on assistant responses
trainer = train_on_responses_only(
    trainer,
    instruction_part = "<start_of_turn>user\n",
    response_part    = "<start_of_turn>model\n",
)
```
